```python
import numpy as np
import freud
import gsd.hoomd

### input parameters
N = 100 # number of particle in a box
phi_p = 0.2 # packing fraction. This initialization is for diulte system. ~0.4 or less recommended
D = 1.0 # particle diameter
r_factor = 1.0 # how much overlap is allowed (1: hard, no overlap)


r_min = D*r_factor
v_0 = 4*np.pi/3*(D/2)**3 # volume of a single particle
L = ((N*v_0)/phi_p)**(1/3) # cubic box length

fbox = freud.box.Box.cube(L) # create freud box to check overlap
pos = np.random.rand(1,3)*L - L/2
# add a particle at a time after overlap check
while len(pos) < N:
    aq = freud.locality.AABBQuery(fbox, pos)
    pos_trial = np.random.rand(1,3)*L - L/2
    if len(list(aq.query(pos_trial, dict(r_max=r_min)))) == 0:
        pos = np.append(pos,pos_trial, axis=0)

snap = gsd.hoomd.Snapshot()
snap.particles.N = N
snap.particles.types = ['A']
snap.particles.typeid = np.zeros(N)
snap.particles.position = pos
snap.particles.diameter = D*np.ones(N)
snap.configuration.box = [fbox.Lx, fbox.Ly, fbox.Lz, fbox.xy, fbox.xz, fbox.yz]
init_gsd = gsd.hoomd.open(name='init.gsd', mode='wb')
init_gsd.append(snap)
#hoomd.init.read_gsd('init_gsd')
```
